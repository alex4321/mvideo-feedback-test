# Загрузка модели

```python
import pickle

with open("model.pkl", "rb") as src:
    model = pickle.load(src)
    
model
```

# Работа с моделью

Загрузим X_train и отберем 50 случайных комментариев:

```python
import pandas
import random

comments = list(pandas.read_csv("X_train.csv")["comment"])
random.shuffle(comments)
comments = comments[:50]
comments
```

Подсчитаем предсказание модели для этих комментариев:

```python
prediction = model.predict(comments)
prediction
```

И покажем таблицу с комментарием и предсказанием

```python
pandas.set_option('display.max_colwidth', -1)
pandas.DataFrame({
    "comment": comments,
    "prediction": prediction
})
```
